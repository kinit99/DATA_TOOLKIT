**QUESTION 01.** Using the Numpy function, generate an array of wRR evenly spaced numPers Petween w and wR and Reshape that wD array into a 2D array ?

```python
#ANSWER 01.
import numpy as np

# Define variables
w = 0  # Starting value
wR = 10  # Ending value
numPers = 20  # Number of evenly spaced numbers
wD = 4  # Number of rows for reshaping (modify this based on desired shape)

# Generate an array of evenly spaced numbers between w and wR
arr = np.linspace(w, wR, numPers)

# Reshape the array into a 2D array with wD rows
reshaped_arr = arr.reshape(wD, -1)  # -1 will automatically calculate the correct number of columns

print(reshaped_arr)
```

**QUESTION 04.**Create a NumPy array with random integers Between 1 and 10 of shape (5,6 )) After creating the array perform the following operations:

 b)Extract all odd integers from array.


```python
#ANSWER 04.
import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
arr = np.random.randint(1, 11, (5, 6))

# a) Extract all even integers from the array
even_integers = arr[arr % 2 == 0]

# b) Extract all odd integers from the array
odd_integers = arr[arr % 2 != 0]

print("Original array:\n", arr)
print("Even integers:\n", even_integers)
print("Odd integers:\n", odd_integers)
```

**QUESTION 13.**Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same Plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the Plot as 'Temperature and Humidity Over Time'.

```python
# ANSWER 13.
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')

# Create random temperature and humidity data
temperature = np.random.uniform(15, 30, size=30)  # Random temperatures between 15 and 30 degrees
humidity = np.random.uniform(40, 90, size=30)     # Random humidity between 40% and 90%

# Create a DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Plot the 'Temperature' and 'Humidity' with two different y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# a) Plot temperature (left y-axis)
ax1.plot(data['Date'], data['Temperature'], color='r', label='Temperature')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create a second y-axis for humidity
ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Humidity'], color='b', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# b) Label the x-axis as 'Date'
ax1.set_xlabel('Date')

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

**QUESTION 14.** Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line Plot representing the normal distribution's Probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the Plot as 'Histogram with PDF Overlay'.

```python
# ANSWER 14.
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# a) Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# b) Plot the histogram with 30 bins
plt.figure(figsize=(10, 6))
count, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# c) Overlay the normal distribution's PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, pdf, 'r-', label='Normal PDF')

# d) Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# e) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

**QUESTION 16.**Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
# ANSWER 16.
# Import necessary libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays with 100 points each
np.random.seed(42)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine which quadrant each point belongs to
def get_quadrant(x_val, y_val):
    if x_val >= 0 and y_val >= 0:
        return 'First Quadrant'
    elif x_val < 0 and y_val >= 0:
        return 'Second Quadrant'
    elif x_val < 0 and y_val < 0:
        return 'Third Quadrant'
    else:
        return 'Fourth Quadrant'

# Apply the quadrant function to the data
quadrants = [get_quadrant(x_val, y_val) for x_val, y_val in zip(x, y)]

# Create a Seaborn scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x=x, y=y, hue=quadrants, palette='Set1', s=100)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()
```

**QUESTION 18.** Using Bokeh, Generate a bar chart of randomly Generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'


```python
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Generate random categorical data
categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Define colors based on values
# Moved this calculation to be done on the DataFrame
data['colors'] = ['#%02x%02x%02x' % (int(value*2.5), 100, 150) for value in values]

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a bar chart
p = figure(x_range=data['categories'], title='Random Categorical Bar Chart',
           tools='hover', tooltips='@values', height=400)

# Referencing the color column from the ColumnDataSource
p.vbar(x='categories', top='values', source=source, width=0.9, color='colors')

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'
p.xgrid.grid_line_color = None

# Show the plot
show(p)
```

**QUESTION 19.** Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

```python
#ANSWER 19.
import plotly.graph_objects as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100) * 10  # Random values

# Create a line plot
fig = go.Figure()

# Add a trace for the line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Label the axes
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

# Show the plot
fig.show()
```

**QUESTION 20.** Using Plotly, create an interactive pie chart of randomly Generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.
